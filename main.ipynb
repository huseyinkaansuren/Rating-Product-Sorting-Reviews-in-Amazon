## Step 2: Calculate the Weighted Score Average by Date.

```python
def time_based_weighted_average(df, w1 = 30, w2 = 26, w3 = 24, w4 = 20):
  return df.loc[df["day_diff"] <= 30, "overall"].mean() * w1/100 +\
          df.loc[(df["day_diff"]> 30) & (df["day_diff"] <= 90), "overall"].mean() * w2/100 +\
            df.loc[(df["day_diff"]> 90) & (df["day_diff"] <= 180), "overall"].mean() * w3/100 +\
              df.loc[(df["day_diff"] > 180), "overall"].mean() * w4/100
```

```python
time_based_weighted_average(df)# Date-weighted score average as a result of a calculation that more weight will be given to recent history
```

## Step 2: Calculate score_pos_neg_diff, score_average_rating and wilson_lower_bound Scores and Add them to the Data

To calculate score_pos_neg_diff, score_average_rating and wilson_lower_bound scores, define the functions **score_pos_neg_diff**, **score_average_rating** and **wilson_lower_bound**.

* Create scores according to score_pos_neg_diff. Next; Save it in df with the name score_pos_neg_diff

* Create scores according to score_average_rating. Then save it in df with the name score_average_rating.

* Create scores according to wilson_lower_bound. Then save it in df with the name wilson_lower_bound.

```python
import math
import scipy.stats as st
```

```python
def score_pos_neg_diff(like, dislike):#Comment score according to like, dislike difference
  return like - dislike
```

```python
def score_average_rating(like, dislike):# average score of the comment
  if like + dislike == 0:
    return 0
  return like / (like + dislike)
```

```python
def wilson_lower_bound(like, dislike, confidence = 0.95):# Scoring with Bernoulli probability method
  n = like + dislike
  if n == 0:
    return 0
  z = st.norm.ppf(1 - (1 - confidence) / 2)
  phat = 1.0 * like / n
  return (phat + z * z / (2 * n) - z * math.sqrt((phat * (1 - phat) + z * z / (4 * n)) / n)) / (1 + z * z / n)
```

```python
df["score_pos_neg_diff"] = df.apply(lambda x: score_pos_neg_diff(x["helpful_yes"], x["helpful_no"]), axis = 1)
df["score_average_rating"] = df.apply(lambda x: score_average_rating(x["helpful_yes"], x["helpful_no"]), axis = 1)
df["wilson_lower_bound"] = df.apply(lambda x: wilson_lower_bound(x["helpful_yes"], x["helpful_no"]), axis = 1)
```

```python
df["days_scored"] = pd.qcut(df["day_diff"], 5, labels = [5, 4, 3, 2, 1])
df["days_scored"] = df["days_scored"].astype(int)
```

```python
min_value = df["days_scored"].min()
max_value = df["days_scored"].max()
def custom_scale(value):
  return 1 - (value - min_value) / (max_value- min_value)
```

```python
df["reverse_scaled_days"]  = df["days_scored"].apply(custom_scale)
```

```python
df["time_wilson_weighted"] = df["reverse_scaled_days"] * 20 / 100 + df["wilson_lower_bound"] * 80 / 100
```
